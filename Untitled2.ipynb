```python
# IBM Employee Data Analysis (Only IBM Dataset)
# Google Colab Compatible Version

# 📦 Install necessary packages (Uncomment if running in Colab)
!pip install pandas matplotlib seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io
from google.colab import files

# 📁 Upload IBM dataset
uploaded = files.upload()

# 📄 Dynamically read IBM dataset (match filename that contains 'ibm')
ibm_filename = next((name for name in uploaded if 'ibm' in name.lower()), None)
if not ibm_filename:
    raise ValueError("IBM dataset file not found in upload.")

ibm_df = pd.read_csv(io.BytesIO(uploaded[ibm_filename]))

# 🧹 Basic Cleaning
ibm_df.dropna(inplace=True)

# 📊 Summary Statistics
print("\n--- IBM Employee Dataset Summary ---")
print(ibm_df.describe())

# 📈 Gender Distribution
if 'Gender' in ibm_df.columns:
    sns.countplot(x='Gender', data=ibm_df)
    plt.title("Gender Distribution")
    plt.show()

# 📊 Attrition by Department
if 'Department' in ibm_df.columns and 'Attrition' in ibm_df.columns:
    sns.countplot(data=ibm_df, x='Department', hue='Attrition')
    plt.title("Attrition by Department")
    plt.xticks(rotation=45)
    plt.show()

# 🧘‍♀ Work-Life Balance vs Job Satisfaction
if 'WorkLifeBalance' in ibm_df.columns and 'JobSatisfaction' in ibm_df.columns:
    sns.boxplot(data=ibm_df, x='WorkLifeBalance', y='JobSatisfaction')
    plt.title("Work-Life Balance vs Job Satisfaction")
    plt.show()

# 💸 Monthly Income Distribution
if 'MonthlyIncome' in ibm_df.columns:
    sns.histplot(ibm_df['MonthlyIncome'], bins=30, kde=True)
    plt.title("Monthly Income Distribution")
    plt.show()

# 📉 Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(ibm_df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# ✅ Save Cleaned Dataset
ibm_df.to_csv("cleaned_ibm.csv", index=False)
print("\nCleaned IBM dataset saved to cleaned_ibm.csv")
```
